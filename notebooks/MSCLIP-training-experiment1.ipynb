```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('experiment1/iconclass-b16-yfcc-msclips_epochs10_as1_start-lr10e-4_lr-schedCosineAnnealingWarmRestarts.pkl')
df
```

```python
def plot_train(file):
    
    df = pd.read_pickle(file)
    
    fig, ax1 = plt.subplots()
    
    ax2 = ax1.twinx()  
    ax3 = ax1.twinx()
    #
    ax3.spines['right'].set_position(('outward', 60))
    
    color = 'r'
    ax1.set_xlabel('batch iteration')
    ax1.set_ylabel('lr0', color=color)
    ax1.plot(df['lr0'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    
    color = 'b'
    ax2.set_ylabel('trloss', color=color)  
    ax2.plot(df['trloss'].rolling(100).mean(), color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
    #color = 'g'
    #ax1.set_xlabel('batch iteration')
    #ax3.set_ylabel('elapsed', color=color)  
    #ax3.plot(df['elapsed'].cumsum(), color=color)
    #ax3.tick_params(axis='y', labelcolor=color)
    
    color = 'k'
    ax3.set_ylabel('update_counter', color=color)  
    ax3.plot(df['update_counter'], color=color)
    ax3.tick_params(axis='y', labelcolor=color)
    

    fig.tight_layout()  
    plt.show()
```

# CosineAnnealingWarmRestarts

## batch size 64

```python
plot_train('experiment1/iconclass-b16-yfcc-msclips_epochs10_as1_start-lr10e-4_lr-schedCosineAnnealingWarmRestarts.pkl')
```
