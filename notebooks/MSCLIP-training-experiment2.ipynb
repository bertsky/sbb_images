```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('experiment2/iconclass-b16-yfcc-msclips_epochs10_as2_start-lr5e-4_lr-schedCosineAnnealingWarmRestarts.pkl')
df
```

```python
def plot_train(file):
    
    df = pd.read_pickle(file)
    
    base_size=10
    fig, ax1 = plt.subplots(figsize=(base_size,int(34/55*base_size)))
    
    ax2 = ax1.twinx()  
    ax3 = ax1.twinx()
    ax4 = ax1.twinx()
    
    ax3.spines['right'].set_position(('outward', 60))
    ax4.spines['right'].set_position(('outward', 120))
    
    color = 'r'
    ax1.set_xlabel('batch iteration')
    ax1.set_ylabel('lr0', color=color)
    ax1.plot(df['lr0'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    
    color = 'b'
    ax2.set_ylabel('trloss', color=color)  
    ax2.plot(df['trloss'].rolling(100).mean(), color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
    color = 'g'
    ax3.set_ylabel('elapsed(hr)', color=color)  
    ax3.plot(df['elapsed']/3600, color=color)
    ax3.tick_params(axis='y', labelcolor=color)
    
    color = 'k'
    ax4.set_ylabel('#updates', color=color)  
    ax4.plot(df['update_counter'], 'k-.')
    ax4.tick_params(axis='y', labelcolor=color)
    

    fig.tight_layout()  
    plt.show()
```

# CosineAnnealingWarmRestarts

## batch size 64

```python
plot_train('experiment2/iconclass-b16-yfcc-msclips_epochs10_as1_start-lr5e-4_lr-schedCosineAnnealingWarmRestarts.pkl')
```
